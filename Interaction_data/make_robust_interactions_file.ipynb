```python
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
from matplotlib.patches import Patch

### charge all interaction data
neg_re2g={}
with open("/home/mouren/Data/interaction/control/neg_encode_re2g_overlap_raw") as file:
    for line in file: 
        if line.strip().split('\t')[7] != "NA":
            if line.strip().split('\t')[3] not in neg_re2g:
                neg_re2g[line.strip().split('\t')[3]] = [line.strip().split('\t')[7]]    
            else:
                neg_re2g[line.strip().split('\t')[3]].append(line.strip().split('\t')[7])

neg_hic = {}
with open("/home/mouren/Data/interaction/control/promHiC_allctrlneg_interactions.tsv") as file:
    for line in file: 
        if line.strip().split('\t')[3] not in neg_hic:
            neg_hic[line.strip().split('\t')[3]] = []
        for i in line.strip().split('\t')[14].split(","):
            if i != "NA":
                neg_hic[line.strip().split('\t')[3]].append(i)


neg_gtex = {}
with open("/home/mouren/Data/interaction/control/ctrlneg_ovlp_gtex_eqtls.tsv") as file:
    for line in file: 
        if line.strip().split('\t')[10].split(".")[0] != "NA":
            if line.strip().split('\t')[3] not in neg_gtex:
                neg_gtex[line.strip().split('\t')[3]] = [line.strip().split('\t')[10].split(".")[0]]    
            else:
                neg_gtex[line.strip().split('\t')[3]].append(line.strip().split('\t')[10].split(".")[0])

pos_re2g={}
with open("/home/mouren/Data/interaction/control/pos_encode_re2g_overlap_raw") as file:
    for line in file: 
        if line.strip().split('\t')[7] != "NA":
            if line.strip().split('\t')[3] not in pos_re2g:
                pos_re2g[line.strip().split('\t')[3]] = [line.strip().split('\t')[7]]    
            else:
                pos_re2g[line.strip().split('\t')[3]].append(line.strip().split('\t')[7])

pos_hic = {}
with open("/home/mouren/Data/interaction/control/promHiC_allctrlpos_interactions.tsv") as file:
    for line in file: 
        if line.strip().split('\t')[3] not in pos_hic:
            pos_hic[line.strip().split('\t')[3]] = []
        for i in line.strip().split('\t')[14].split(","):
            if i != "NA":
                pos_hic[line.strip().split('\t')[3]].append(i)

pos_gtex = {}
with open("/home/mouren/Data/interaction/control/ctrlpos_ovlp_gtex_eqtls.tsv") as file:
    for line in file: 
        if line.strip().split('\t')[8].split(".")[0] != "NA":
            if line.strip().split('\t')[3] not in pos_gtex:
                pos_gtex[line.strip().split('\t')[3]] = [line.strip().split('\t')[8].split(".")[0]]    
            else:
                pos_gtex[line.strip().split('\t')[3]].append(line.strip().split('\t')[8].split(".")[0])
```

```python
#neg
content=[]
with open("/home/mouren/Data/final_files_tokeep/control/control_neg_NoTF_NoTSS_TES_prom.tsv") as file:
    for line in file: 
        try:
            set_re2g = set(neg_re2g[line.strip().split("\t")[3]])
        except KeyError:
            set_re2g = set()
        try:
            set_hic = set(neg_hic[line.strip().split("\t")[3]])
        except KeyError:
            set_hic = set()
        try:
            set_gtex = set(neg_gtex[line.strip().split("\t")[3]])
        except KeyError:
            set_gtex = set()
        
        values_in_all_three = set_re2g & set_hic & set_gtex 
        values_in_AB = (set_re2g & set_hic) - values_in_all_three
        values_in_BC = (set_hic & set_gtex) - values_in_all_three
        values_in_AC = (set_re2g & set_gtex) - values_in_all_three

        for i in values_in_all_three:
            content.append([line.strip().split("\t")[3],i,"All"])
        for i in values_in_AB:
            content.append([line.strip().split("\t")[3],i,"ENCODE-rE2G_Pcap-HiC"])
        for i in values_in_BC:
            content.append([line.strip().split("\t")[3],i,"Pcap-HiC_GTEX-eQTL"])
        for i in values_in_AC:
            content.append([line.strip().split("\t")[3],i,"ENCODE-rE2G_GTEX-eQTL"])
                   
df = pd.DataFrame(content)
df.to_csv("/home/mouren/Data/interaction/control/robust_associations_ctrlneg_genes.txt",sep="\t",header=False,index=False)
```

```python
###Â EE
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
from matplotlib.patches import Patch
from matplotlib_venn import venn2,venn3

encode_exons =  set()
with open('/home/mouren/Data/encode-rE2G/ee_encode_re2g_overlap_raw') as file:
    for line in file:
        encode_exons.add((line.strip().split()[3],line.strip().split()[7]))

hic_exons =  set()
with open('/home/mouren/Data/hiC/promHiC_allEE_interactions.tsv') as file:
    for line in file:
        for i in (line.strip().split()[14]).split(","):
            hic_exons.add((line.strip().split()[3],i))

gtex_exons =  set()
with open('/home/mouren/Data/gtex/ee_overlap_gtexSNP.tsv') as file:
    for line in file:
        gtex_exons.add((line.strip().split()[3],(line.strip().split()[10]).split(".")[0]))


# only_encode_exons = encode_exons - (hic_exons | gtex_exons)
# only_hic_exons = hic_exons - (encode_exons | gtex_exons)
# only_gtex_exons = gtex_exons - (encode_exons | hic_exons)
encode_hic_overlap = encode_exons & hic_exons - gtex_exons
encode_gtex_overlap = encode_exons & gtex_exons - hic_exons
hic_gtex_overlap = hic_exons & gtex_exons - encode_exons
all_three_overlap = encode_exons & hic_exons & gtex_exons

def write_associations_to_file(filename, associations_dict):
    with open(filename, 'w') as file:
        for origin, associations in associations_dict.items():
            for key, value in associations:
                if value != "NA":
                    file.write(f"{key}\t{value}\t{origin}\n")

# Gather the associations along with their origins into a dictionary
associations_dict = {
    "ENCODE-rE2G_Pcap-HiC": encode_hic_overlap,
    "ENCODE-rE2G_GTEX-eQTL": encode_gtex_overlap,
    "Pcap-HiC_GTEX-eQTL": hic_gtex_overlap,
    "All": all_three_overlap
}

filename = "/home/mouren/Data/variants/tcga_MC3/robust_associations_exons_genes.txt"

# Call the function with the filename and the dictionary of associations
write_associations_to_file(filename, associations_dict)
```
