```python
# rescale data
def rescale_data(data):
    sorted_values = sorted(data.values())
    x_original = np.linspace(0, 100, len(sorted_values))  # Generate rescaled x-axis (0 to 100)
    return x_original, sorted_values

# Rescale each data series
x1, y1 = rescale_data(hg_ee)
x2, y2 = rescale_data(hg_ee_shuf)
x3, y3 = rescale_data(hg_pos)
x4, y4 = rescale_data(hg_pos_shuf)

x5, y5 = rescale_data(mm_ee)
x6, y6 = rescale_data(mm_ee_shuf)
x7, y7 = rescale_data(mm_pos)
x8, y8 = rescale_data(mm_pos_shuf)

x9, y9 = rescale_data(dm_ee)
x10, y10 = rescale_data(dm_ee_shuf)
x11, y11 = rescale_data(dm_pos)
x12, y12 = rescale_data(dm_pos_shuf)

x13, y13 = rescale_data(tair_ee)
x14, y14 = rescale_data(tair_ee_shuf)
x15, y15 = rescale_data(tair_pos)
x16, y16 = rescale_data(tair_pos_shuf)

print(np.mean(y1),np.median(y1))
print(np.mean(y2),np.median(y2))

print(np.mean(y9),np.median(y9))
print(np.mean(y10),np.median(y10))
```
