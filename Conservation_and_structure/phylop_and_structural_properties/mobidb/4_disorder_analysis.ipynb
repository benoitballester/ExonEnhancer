```python
import json 

def check_number_in_range(element, number):
    ranges = genes_disorder.get(element, [])
    for range_set in ranges:
        if int(range_set[0]) <= int(number) <= int(range_set[1]):
            return True
    return False
```

```python
# HSAP EE
output_file = "/home/mouren/Data/mobidb/hsap/ee_disorder_perc_hg38.tsv"

with open('/home/mouren/Data/mobidb/hsap/genes_disorder_sorted_mobiDB_hg38.json', 'r') as f:
    genes_disorder = json.load(f)

transcripts_pep = {}
with open("/home/mouren/Data/mobidb/hsap/mobi_entry_full_aa_seq_hg.tsv") as file:
    for line in file:
        transcripts_pep[line.strip().split()[0]] = line.strip().split()[1]

dic_pep_ee = {}
with open("/home/mouren/Data/final_files_tokeep/aa_sequences/sequences/ee_aa_seqs_hg38.txt") as file:
    for line in file:      
        if line.strip().split()[1] != "false":
            dic_pep_ee[line.strip().split()[0]] = line.strip().split()[1]

dic_disorder_total = {}
with open("/home/mouren/Data/final_files_tokeep/notes/gene_names/mobidb_id_to_dl/hg_ee_uniprotID") as file:
    for line in file:
  
        try:
            transcript_seq = transcripts_pep[line.strip().split("\t")[1]]
        except KeyError: # the mobidb id didnt give any disorder result
            continue
      
        if line.strip().split("\t")[0] in dic_pep_ee:
            start_index = transcript_seq.find(dic_pep_ee[line.strip().split("\t")[0]]) #get postion of exon AA sequence in gene seq

            if start_index != -1:
                end_index = start_index + len(dic_pep_ee[line.strip().split("\t")[0]])
                start_index += 1 #+1 because in python it stars counting at zero
                end_index += 1 #+1 because not included when looping with range

                cpt_true = 0
                for aa_number in range(start_index,end_index):
                    check = check_number_in_range(line.strip().split("\t")[1], aa_number)
                    if check == True:
                        cpt_true += 1
                
                res = (cpt_true*100)/(end_index-start_index)

                dic_disorder_total[line.strip().split("\t")[0]] = res

with open(output_file, 'w') as f:
    for key, value in dic_disorder_total.items():
        f.write(f"{key}\t{value}\n")
```
