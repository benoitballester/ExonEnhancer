```python
###Â prepare files of results
import pandas as pd 

dic_clade = {}
with open("/home/mouren/Data/age_genes/clades_names_Trigos2017.csv") as f:
    first = True
    for line in f:
        if first is True:
            first = False
            continue
        else:
            dic_clade[line.strip().split("\t")[0]] = line.strip().split("\t")[1]

dic_age={}
with open("/home/mouren/Data/age_genes/age_genes_Trigos2017.txt") as file:
    for line in file:            
        dic_age[line.strip().split(",")[0].upper()] = line.strip().split(",")[2]

dic_transcript_to_genes = {}
with open("/home/mouren/Data/final_files_tokeep/notes/list_transript_gene_id_gencodeV41_all_proteincoding.tsv") as file:
    for line in file:            
        dic_transcript_to_genes[line.strip().split("\t")[0]] = (line.strip().split("\t")[1]).split(".")[0]

dic_genes_to_name = {}
with open("/home/mouren/Data/final_files_tokeep/notes/ensemblID_geneID_dic.tsv") as file:
    first = True
    for line in file:   
        if first :
            first = False
            continue
        splitted = line.strip().split("\t")
        if len(splitted) == 1:
            continue
        if line.strip().split("\t")[0] not in dic_genes_to_name:
            dic_genes_to_name[line.strip().split("\t")[0]] = [line.strip().split("\t")[1].upper()]
            if len(splitted) > 2:
                dic_genes_to_name[line.strip().split("\t")[0]].append(line.strip().split("\t")[2].upper())
        else:
            dic_genes_to_name[line.strip().split("\t")[0]].append(line.strip().split("\t")[2].upper())

content = [["Exon", "Ensembl_GeneID", "GeneID", "Clade_number", "Clade", "Category"]]
with open("/home/mouren/Data/final_files_tokeep/final_catalogs/robust_catalog_hsap.tsv") as file:
    for line in file:    
        gene = dic_transcript_to_genes[(line.strip().split("\t")[3]).split("_")[0]]
        
        try:
            name = dic_genes_to_name[gene]
        except KeyError:
            entry = [line.strip().split("\t")[3], gene, "NA", "NA", "NA", "NA"]
            content.append(entry)
            continue
        
        age = "NA"
        name2 = "NA"
        for i in name:
            try:
                age = dic_age[i]
                name2 = i
                break
            except KeyError:
                continue

        try:
            clade = dic_clade[age]
        except KeyError:
            clade = "NA"

        cat = "NA"
        if age != "NA":
            if int(age) <= 3:
                cat = "UC"
            elif int(age) <= 9:
                cat = "EM"
            elif int(age) <= 16:
                cat = "MM"

        entry = [line.strip().split("\t")[3], gene, name2, age, clade, cat]
        content.append(entry)

df = pd.DataFrame(content)
df.to_csv("/home/mouren/Data/age_genes/ee_age_genes.tsv",sep="\t",header=False,index=False)


content = [["Exon", "Ensembl_GeneID", "GeneID", "Clade_number", "Clade", "Category"]]
with open("/home/mouren/Data/final_files_tokeep/control/control_neg_NoTF_NoTSS_TES_prom.tsv") as file:
    for line in file:    
        gene = dic_transcript_to_genes[(line.strip().split("\t")[3]).split("_")[0]]
        
        try:
            name = dic_genes_to_name[gene]
        except KeyError:
            entry = [line.strip().split("\t")[3], gene, "NA", "NA", "NA", "NA"]
            content.append(entry)
            continue
        
        age = "NA"
        name2 = "NA"
        for i in name:
            try:
                age = dic_age[i]
                name2 = i
                break
            except KeyError:
                continue

        try:
            clade = dic_clade[age]
        except KeyError:
            clade = "NA"

        cat = "NA"
        if age != "NA":
            if int(age) <= 3:
                cat = "UC"
            elif int(age) <= 9:
                cat = "EM"
            elif int(age) <= 16:
                cat = "MM"

        entry = [line.strip().split("\t")[3], gene, name2, age, clade, cat]
        content.append(entry)

df2 = pd.DataFrame(content)
df2.to_csv("/home/mouren/Data/age_genes/ctrlNeg_age_genes.tsv",sep="\t",header=False,index=False)
```
