```python
import matplotlib.pyplot as plt
import seaborn as sns 
import matplotlib 
matplotlib.rcParams['pdf.fonttype'] = 42

color1 = sns.color_palette()
color2=sns.color_palette('pastel')

# Data for Group A and Group B
ee_values = [27.742]
all_values = [70.751]
not_mapped = [0.897]
not_lifted = [0.608]

# Figure size
plt.figure(figsize=(2, 5))  # Adjusted for better visibility of all groups

# Plotting
plt.bar("Hg38 to Mm39",ee_values, label='Matching EE', color=color1[3], edgecolor='black', linewidth=0.5)
plt.bar("Hg38 to Mm39",all_values, bottom=ee_values, label='Matching coding exons', color=color1[5], edgecolor='black', linewidth=0.5)
plt.bar("Hg38 to Mm39",not_mapped, bottom=[27.742+70.751], label='No overlap', color=color1[7], edgecolor='black', linewidth=0.5)
plt.bar("Hg38 to Mm39",not_lifted, bottom=[27.742+70.751+0.897], label='Not lifted', color=color1[7], hatch='///', edgecolor='black', linewidth=0.5)

# Overall labels, title and legend
plt.ylabel('Percentage')
plt.title('Conservation of EE from Hg38 to Mm39')
plt.xlabel(xlabel=None)  # Assuming these are different metrics being compared
plt.legend(bbox_to_anchor=(1.01, 1),fontsize=8)

plt.ylim(0, 100)  # Adjusted to accommodate the stacked bars' heights

plt.savefig("/home/mouren/Images/tls1/conservation/EE_between_hg38_mm39_lastz.pdf", format="pdf", bbox_inches="tight")

plt.show()
```
