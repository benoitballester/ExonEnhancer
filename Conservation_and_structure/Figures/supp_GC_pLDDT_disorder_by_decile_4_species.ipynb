```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from statannot import add_stat_annotation
import numpy as np 

# Use Type 42 fonts for PDFs
matplotlib.rcParams['pdf.fonttype'] = 42

# Function to pad lists with np.nan
def pad_list(lst, length):
    return lst + [np.nan] * (length - len(lst))

# Load your data
df1 = pd.read_csv("/home/mouren/Data/final_files_tokeep/notes/species_cara_bilan/bilan_ee_caracteristique_hg38.tsv", sep='\t', header=0)
df2 = pd.read_csv("/home/mouren/Data/final_files_tokeep/notes/species_cara_bilan/bilan_ee_caracteristique_mm39.tsv", sep='\t', header=0)
df3 = pd.read_csv("/home/mouren/Data/final_files_tokeep/notes/species_cara_bilan/bilan_ee_caracteristique_dm6.tsv", sep='\t', header=0)
df4 = pd.read_csv("/home/mouren/Data/final_files_tokeep/notes/species_cara_bilan/bilan_ee_caracteristique_tair10.tsv", sep='\t', header=0)

max_len = max(len(neg_gc_hg), len(neg_plddt_hg), len(neg_disorder_hg))
df_control = pd.DataFrame({"GC%": pad_list(neg_gc_hg, max_len),"Mean_alphafold_pLDDT": pad_list(neg_plddt_hg, max_len),"MobiDB_disorder%": pad_list(neg_disorder_hg, max_len),"GroupByTF": ["Ctrl-"] * max_len})
df_control['GroupByTF'] = "Ctrl-"
df1_2 = pd.concat([df1, df_control])

max_len = max(len(neg_gc_mm), len(neg_plddt_mm), len(neg_disorder_mm))
df_control = pd.DataFrame({"GC%": pad_list(neg_gc_mm, max_len),"Mean_alphafold_pLDDT": pad_list(neg_plddt_mm, max_len),"MobiDB_disorder%": pad_list(neg_disorder_mm, max_len),"GroupByTF": ["Ctrl-"] * max_len})
df_control['GroupByTF'] = "Ctrl-"
df2_2 = pd.concat([df2, df_control])

max_len = max(len(neg_gc_dm), len(neg_plddt_dm), len(neg_disorder_dm))
df_control = pd.DataFrame({"GC%": pad_list(neg_gc_dm, max_len),"Mean_alphafold_pLDDT": pad_list(neg_plddt_dm, max_len),"MobiDB_disorder%": pad_list(neg_disorder_dm, max_len),"GroupByTF": ["Ctrl-"] * max_len})
df_control['GroupByTF'] = "Ctrl-"
df3_2 = pd.concat([df3, df_control])

max_len = max(len(neg_gc_tair), len(neg_plddt_tair), len(neg_disorder_tair))
df_control = pd.DataFrame({"GC%": pad_list(neg_gc_tair, max_len),"Mean_alphafold_pLDDT": pad_list(neg_plddt_tair, max_len),"MobiDB_disorder%": pad_list(neg_disorder_tair, max_len),"GroupByTF": ["Ctrl-"] * max_len})
df_control['GroupByTF'] = "Ctrl-"
df4_2 = pd.concat([df4, df_control])
```

```python
# Define the columns and DataFrames
columns_of_interest = ['GC%', 'Mean_alphafold_pLDDT', 'MobiDB_disorder%']
dfs = [df1_2, df2_2, df3_2, df4_2]
colors = ['red', 'silver', 'orange', 'green']

# Create the figure
fig, axes = plt.subplots(4, 3, figsize=(8, 12), sharey=True, sharex=True)

# Loop over each feature (column) and each dataset (row)
for col_idx, column in enumerate(columns_of_interest):
    for row_idx, (df, color) in enumerate(zip(dfs, colors)):
        # Prepare data
        data_tmp = df[['GroupByTF', column]].dropna()
        #data_tmp["GroupByTF"] = pd.to_numeric(data_tmp["GroupByTF"])
        data_tmp[column] = pd.to_numeric(data_tmp[column])

        ax = axes[row_idx, col_idx]
        
        # Create boxplot
        sns.boxplot(
            data=data_tmp,
            x='GroupByTF', 
            y=column, 
            palette=['dimgrey'] + [color]*10,  # first box gray, others red
            ax=ax,
            showfliers=True,    
            linewidth=1.5,       
            whiskerprops={'linestyle': '--'},
            flierprops={
                'marker': 'o',
                'markerfacecolor': 'none',
                'markeredgecolor': 'black',
                'markersize': 5
            },
            order=['Ctrl-'] + list(range(1, 11))
        )
        
        # Set plot formatting
        if row_idx == 0:
            ax.set_title(None)
        ax.set_xlabel(None)
        ax.set_ylabel(None)
        ax.tick_params(axis='x', labelsize=8)
        ax.tick_params(axis='y', labelsize=8)
        ax.set_ylim(-1, 101)
        
        if column=="GC%":
            add_stat_annotation(ax,data=data_tmp,x='GroupByTF',y=column,box_pairs=[(1, 10)],order=['Ctrl-'] + list(range(1, 11)),pvalues=[0],test=None,perform_stat_test=False,text_format='star',loc='outside')
        else:            
            # Add Kruskal-Wallis test for the defined pairs
            add_stat_annotation(
                ax, data=data_tmp, x='GroupByTF', y=column,
                box_pairs=[(1, 10)],
                order=['Ctrl-'] + list(range(1, 11)),
                test='Kruskal',  # Kruskal-Wallis test
                text_format='star',
                loc='outside',
            )#verbose=0)
        
fig.text(0.205, 1.01, "GC content", ha='center', fontsize=12)
fig.text(0.515, 1.01, "AlphaFold mean pLDDT", ha='center', fontsize=12)
fig.text(0.84, 1.01, "MobiDB %protein disorder", ha='center', fontsize=12)

plt.tight_layout()
plt.savefig("/home/mouren/Images/tls1/supplementary_figures/supp_fig7/GC_pLDTT_disorder_by_decile_all_species.pdf", format="pdf", bbox_inches="tight", dpi=450)
plt.savefig("/home/mouren/Images/tls1/supplementary_figures/supp_fig7/GC_pLDTT_disorder_by_decile_all_species.png", format="png", bbox_inches="tight", dpi=450)
plt.show()
```
