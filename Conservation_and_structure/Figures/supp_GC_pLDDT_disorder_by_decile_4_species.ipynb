```python
# Define the columns and DataFrames
# Define the columns and DataFrames
columns_of_interest = ['GC%', 'Mean_alphafold_pLDDT', 'MobiDB_disorder%']
dfs = [df1_2, df2_2, df3_2, df4_2]
colors = ['#1175BC', 'silver', 'orange', 'green']

# Create the figure
fig, axes = plt.subplots(4, 3, figsize=(8, 12), sharey=True, sharex=True)

order = ['Ctrl-'] + list(range(1, 11))  # <--- ADD THIS ONCE

plist_gc = [0,1.6292611181141932e-250,5.086824604907018e-98,3.6361312794011113e-66]    
plist_alpha = [1.3484071019928463e-26,1.2270106573497417e-29,6.269526039649894e-83,0.03234742804249917]
plist_mobi = [1.6772297971016067e-50,1.1622236271961096e-47,9.863465542693305e-116,9.038068139446888e-08]     

for col_idx, column in enumerate(columns_of_interest):
    for row_idx, (df, color) in enumerate(zip(dfs, colors)):

        data_tmp = df[['GroupByTF', column]].dropna()
        data_tmp[column] = pd.to_numeric(data_tmp[column])

        ax = axes[row_idx, col_idx]

        sns.boxplot(
            data=data_tmp,
            x='GroupByTF',
            y=column,
            palette=['dimgrey'] + [color]*10,
            ax=ax,
            showfliers=True,
            linewidth=1.5,
            whiskerprops={'linestyle': '--'},
            flierprops={
                'marker': 'o',
                'markerfacecolor': 'none',
                'markeredgecolor': 'black',
                'markersize': 5
            },
            order=order
        )

        # ---- ADD: n per box annotation ----
        n_per_group = (
            data_tmp.groupby('GroupByTF')[column]
            .size()
            .reindex(order, fill_value=0)
        )
        y_text = -10  # adjust if needed
        for i, grp in enumerate(order):
            ax.text(
                i, y_text, f"n={int(n_per_group.loc[grp])}",
                ha="center", va="top",
                fontsize=7, rotation=45,
                clip_on=False
            )
        # -----------------------------------

        if row_idx == 0:
            ax.set_title(None)
        ax.set_xlabel(None)
        ax.set_ylabel(None)
        ax.tick_params(axis='x', labelsize=8)
        ax.tick_params(axis='y', labelsize=8)
        ax.set_ylim(-1, 101)


        
        if column=="GC%":
            add_stat_annotation(
                ax,
                data=data_tmp,
                x='GroupByTF',
                y=column,
                box_pairs=[(1, 10)],
                order=['Ctrl-'] + list(range(1, 11)),
                test=None,
                perform_stat_test=False,
                text_format='full',
                pvalues=[plist_gc[0]],
                loc='outside')
            
            plist_gc.pop(0)
        
        elif column=="Mean_alphafold_pLDDT":
            add_stat_annotation(
                ax,
                data=data_tmp,
                x='GroupByTF',
                y=column,
                box_pairs=[(1, 10)],
                order=['Ctrl-'] + list(range(1, 11)),
                test=None,
                perform_stat_test=False,
                text_format='full',
                pvalues=[plist_alpha[0]],
                loc='outside')
            
            plist_alpha.pop(0)
        
        elif column=="MobiDB_disorder%":
            add_stat_annotation(
                    ax,
                    data=data_tmp,
                    x='GroupByTF',
                    y=column,
                    box_pairs=[(1, 10)],
                    order=['Ctrl-'] + list(range(1, 11)),
                    test=None,
                    perform_stat_test=False,
                    text_format='full',
                    pvalues=[plist_mobi[0]],
                    loc='outside')
                
            plist_mobi.pop(0)
                
            
            
fig.text(0.205, 1.01, "GC content", ha='center', fontsize=12)
fig.text(0.515, 1.01, "AlphaFold mean pLDDT", ha='center', fontsize=12)
fig.text(0.84, 1.01, "MobiDB %protein disorder", ha='center', fontsize=12)

plt.tight_layout()
plt.savefig("/home/mouren/Images/supplementary_figures/supp_fig7/GC_pLDTT_disorder_by_decile_all_species.pdf", format="pdf", bbox_inches="tight", dpi=450)
plt.savefig("/home/mouren/Images/supplementary_figures/supp_fig7/GC_pLDTT_disorder_by_decile_all_species.png", format="png", bbox_inches="tight", dpi=450)
plt.show()
```
