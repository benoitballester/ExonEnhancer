```python
### Data and functions

path_files = "/home/mouren/Data/article_repo_data"

def calculate_gc_content(dna_sequence):
    dna_sequence = dna_sequence.upper()

    g_count = dna_sequence.count('G')
    c_count = dna_sequence.count('C')

    total_length = len(dna_sequence)

    if total_length == 0:
        return 0.0
    gc_content_percentage = ((g_count + c_count) / total_length) * 100

    return gc_content_percentage

def calculate_obs_exp_cpg(seq):
    num_c = seq.count('C')
    num_g = seq.count('G')

    # Count observed 'CG' dinucleotides
    obs_cpg = 0
    for i in range(len(seq) - 1):
        if seq[i:i+2] == 'CG':
            obs_cpg += 1
    
    # Calculate expected CpG
    # (avoid dividing by zero if length is 0)
    length = len(seq)
    if length == 0:
        return 0
    
    exp_cpg = (num_c * num_g) / length if length > 0 else 0
    
    # Compute O/E ratio
    oe_ratio = obs_cpg / exp_cpg if exp_cpg != 0 else 0
    return oe_ratio

ee_h=[]
ee_h_isl=[]
with open(path_files+"/Data/sequences_species/hsap/hg38_ee_dna_seqs.tsv") as file:
    for line in file:  
        dna_sequence = line.strip().split("\t")[1]
        gc_content = calculate_gc_content(dna_sequence)
        ee_h.append(gc_content)
        isl_content = calculate_obs_exp_cpg(dna_sequence)
        ee_h_isl.append(isl_content)
neg_h=[]
neg_h_isl=[]
with open(path_files+"/Data/sequences_species/hsap/hg38_neg_dna_seqs.tsv") as file:
    for line in file:  
        dna_sequence = line.strip().split("\t")[1]
        gc_content = calculate_gc_content(dna_sequence)
        neg_h.append(gc_content)
        isl_content = calculate_obs_exp_cpg(dna_sequence)
        neg_h_isl.append(isl_content)
pos_h=[]
pos_h_isl=[]
with open(path_files+"/Data/sequences_species/hsap/hg38_pos_dna_seqs.tsv") as file:
    for line in file:  
        try:
            dna_sequence = line.strip().split("\t")[1]
        except IndexError:
            continue
        gc_content = calculate_gc_content(dna_sequence)
        pos_h.append(gc_content)
        isl_content = calculate_obs_exp_cpg(dna_sequence)
        pos_h_isl.append(isl_content)

ee_m=[]
ee_m_isl=[]
with open(path_files+"/Data/sequences_species/mmus/mm39_ee_dna_seqs.tsv") as file:
    for line in file:  
        dna_sequence = line.strip().split("\t")[1]
        gc_content = calculate_gc_content(dna_sequence)
        ee_m.append(gc_content)
        isl_content = calculate_obs_exp_cpg(dna_sequence)
        ee_m_isl.append(isl_content)
neg_m=[]
neg_m_isl=[]
with open(path_files+"/Data/sequences_species/mmus/mm39_neg_dna_seqs.tsv") as file:
    for line in file:  
        dna_sequence = line.strip().split("\t")[1]
        gc_content = calculate_gc_content(dna_sequence)
        neg_m.append(gc_content)
        isl_content = calculate_obs_exp_cpg(dna_sequence)
        neg_m_isl.append(isl_content)
pos_m=[]
pos_m_isl=[]
with open(path_files+"/Data/sequences_species/mmus/mm39_pos_dna_seqs.tsv") as file:
    for line in file:  
        try:
            dna_sequence = line.strip().split("\t")[1]
        except IndexError:
            continue
        gc_content = calculate_gc_content(dna_sequence)
        pos_m.append(gc_content)
        isl_content = calculate_obs_exp_cpg(dna_sequence)
        pos_m_isl.append(isl_content)

ee_d=[]
ee_d_isl=[]
with open(path_files+"/Data/sequences_species/dmel/dm6_ee_dna_seqs.tsv") as file:
    for line in file:  
        dna_sequence = line.strip().split("\t")[1]
        gc_content = calculate_gc_content(dna_sequence)
        ee_d.append(gc_content)
        isl_content = calculate_obs_exp_cpg(dna_sequence)
        ee_d_isl.append(isl_content)
neg_d=[]
neg_d_isl=[]
with open(path_files+"/Data/sequences_species/dmel/dm6_neg_dna_seqs.tsv") as file:
    for line in file:  
        dna_sequence = line.strip().split("\t")[1]
        gc_content = calculate_gc_content(dna_sequence)
        neg_d.append(gc_content)
        isl_content = calculate_obs_exp_cpg(dna_sequence)
        neg_d_isl.append(isl_content)
pos_d=[]
pos_d_isl=[]
with open(path_files+"/Data/sequences_species/dmel/dm6_pos_dna_seqs.tsv") as file:
    for line in file:  
        try:
            dna_sequence = line.strip().split("\t")[1]
        except IndexError:
            continue
        gc_content = calculate_gc_content(dna_sequence)
        pos_d.append(gc_content)
        isl_content = calculate_obs_exp_cpg(dna_sequence)
        pos_d_isl.append(isl_content)

ee_t=[]
ee_t_isl=[]
with open(path_files+"/Data/sequences_species/thal/tair10_ee_dna_seqs.tsv") as file:
    for line in file:  
        dna_sequence = line.strip().split("\t")[1]
        gc_content = calculate_gc_content(dna_sequence)
        ee_t.append(gc_content)
        isl_content = calculate_obs_exp_cpg(dna_sequence)
        ee_t_isl.append(isl_content)
neg_t=[]
neg_t_isl=[]
with open(path_files+"/Data/sequences_species/thal/tair10_neg_dna_seqs.tsv") as file:
    for line in file:  
        dna_sequence = line.strip().split("\t")[1]
        gc_content = calculate_gc_content(dna_sequence)
        neg_t.append(gc_content)
        isl_content = calculate_obs_exp_cpg(dna_sequence)
        neg_t_isl.append(isl_content)
pos_t=[]
pos_t_isl=[]
with open(path_files+"/Data/sequences_species/thal/tair10_pos_dna_seqs.tsv") as file:
    for line in file:  
        try:
            dna_sequence = line.strip().split("\t")[1]
        except IndexError:
            continue
        gc_content = calculate_gc_content(dna_sequence)
        pos_t.append(gc_content)
        isl_content = calculate_obs_exp_cpg(dna_sequence)
        pos_t_isl.append(isl_content)
```

```python
### filter cpg ratio data
# Calculate IQR boundaries
Q1 = df_isl_exploded['Obs/Exp CpG'].quantile(0.25)
Q3 = df_isl_exploded['Obs/Exp CpG'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the dataset to remove outliers
df_isl_filtered = df_isl_exploded[
    (df_isl_exploded['Obs/Exp CpG'] >= lower_bound) &
    (df_isl_exploded['Obs/Exp CpG'] <= upper_bound)
]
```
