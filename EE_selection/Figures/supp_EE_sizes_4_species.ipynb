```python
### Dataframes for plots
data_all = [
{"Species":"H.sap","Dataset":"All exons","Size (bp)":hg_all},
{"Species":"H.sap","Dataset":"cEEs","Size (bp)":hg_candidates},
{"Species":"H.sap","Dataset":"Ctrl-","Size (bp)":hg_neg},

{"Species":"M.mus","Dataset":"All exons","Size (bp)":mm_all},
{"Species":"M.mus","Dataset":"cEEs","Size (bp)":mm_candidates},
{"Species":"M.mus","Dataset":"Ctrl-","Size (bp)":mm_neg},

{"Species":"D.mel","Dataset":"All exons","Size (bp)":dm_all},
{"Species":"D.mel","Dataset":"cEEs","Size (bp)":dm_candidates},
{"Species":"D.mel","Dataset":"Ctrl-","Size (bp)":dm_neg},

{"Species":"A.tha","Dataset":"All exons","Size (bp)":tair_all},
{"Species":"A.tha","Dataset":"cEEs","Size (bp)":tair_candidates},
{"Species":"A.tha","Dataset":"Ctrl-","Size (bp)":tair_neg},
]
df_all = pd.DataFrame(data_all)
df_all_exploded = df_all.explode('Size (bp)') # Explode 'Nb of exons' lists into separate rows
df_all_exploded['Size (bp)'] = pd.to_numeric(df_all_exploded['Size (bp)']) # Convert 'Nb of exons' to numeric
df_all_exploded['log_Size (bp)'] = np.log10(df_all_exploded['Size (bp)']) # Apply log transformation to the 'Size (bp)' column
```

```python
### PLOT
box_pairs_all = [
    (("H.sap", "All exons"), ("H.sap", "cEEs")),
    (("M.mus", "All exons"), ("M.mus", "cEEs")),
    (("D.mel", "All exons"), ("D.mel", "cEEs")),
    (("A.tha", "All exons"), ("A.tha", "cEEs")),
    (("H.sap", "Ctrl-"), ("H.sap", "cEEs")),
    (("M.mus", "Ctrl-"), ("M.mus", "cEEs")),
    (("D.mel", "Ctrl-"), ("D.mel", "cEEs")),
    (("A.tha", "Ctrl-"), ("A.tha", "cEEs"))
]

##All exons
plt.figure(figsize=(5, 5))
ax = sns.violinplot(data=df_all_exploded, x="Species", y="log_Size (bp)", hue="Dataset", linewidth=1, cut=0)

add_stat_annotation(ax, data=df_all_exploded, x="Species", y="log_Size (bp)", hue="Dataset",
                    box_pairs=box_pairs_all,
                    test='t-test_ind', text_format='simple', loc='inside', verbose=1)

log_ticks = [0, 1, 2, 3, 4, 5]  # These are the log values corresponding to 100, 1000, 10000, etc.
original_ticks = [10**tick for tick in log_ticks]  # Convert log values to the original scale
ax.set_yticks(log_ticks)
ax.set_yticklabels(original_ticks)
ax.set_ylabel('Size (bp)')

ax.set_xlabel(None)

ax.legend(title=None, frameon = False)

plt.tight_layout()

plt.savefig("/home/mouren/Images/tls1/supplementary_figures/supp_fig1/EE_sizes_all_species.pdf", format="pdf", bbox_inches="tight", dpi=450)
plt.savefig("/home/mouren/Images/tls1/supplementary_figures/supp_fig1/EE_sizes_all_species.png", format="png", bbox_inches="tight", dpi=450)

plt.show()
```
