```python
import pandas as pd
import numpy as np

###Â Get dictionnary of number of peaks in biotypes and cells lines 
nb_peak_tf = {}
with open("/home/mouren/Data/final_files_tokeep/tls1/annotation/annot_nb_peaksAll_Remap_TF.tsv") as file:
    for line in file:
        nb_peak_tf[line.strip().split()[1]] = int(line.strip().split()[0])

nb_peak_cellLines = {}
with open("/home/mouren/Data/final_files_tokeep/tls1/annotation/annot_nb_peaksAll_Remap_cellLines.tsv") as file:
    for line in file:
        nb_peak_cellLines[line.strip().split()[1]] = int(line.strip().split()[0])

nb_peak_bio = {}
with open("/home/mouren/Data/final_files_tokeep/tls1/annotation/annot_nb_peaksAll_Remap_biotypeBiotypes.tsv") as file:
    for line in file:
        nb_peak_bio[line.strip().split()[1]] = int(line.strip().split()[0])

### Get bilan all remap overlap on robust exons
content=[]
with open("/home/mouren/Data/final_files_tokeep/tls1/biotype/bilan_robust_exons_OvlpAll_remap.csv") as file:
    for line in file:            
        content.append(line.strip().split())
df = pd.DataFrame(content)
```

```python
### Normalization methodology
# 1: divide the number of tf on an element by the total number of this tf in the remap dataset
# 2: additionate each result for an element
# 3: divide each new value by the result of the addition (to scale the result between 0 and 1)

### Function to call later
def fill_list(lst):
    while len(lst) < 3:
        lst.append(("NA","NA"))
    return lst
```

```python
### TF
content_tf = []
for index, row in df.iterrows():
    somme = 0
    normalized = {}
    
    for tf in range(len(row[2].split(","))):
        normalized[row[2].split(",")[tf]] = int(row[3].split(",")[tf])/nb_peak_tf[row[2].split(",")[tf]]
        somme += int(row[3].split(",")[tf])/nb_peak_tf[row[2].split(",")[tf]]

    for key,value in normalized.items():
        normalized[key] = value/somme

    sorted_normalized = sorted(normalized.items(), key=lambda x: x[1], reverse=True)

    if sorted_normalized[0][1] > 0.5:
        above_50 = "TRUE"
    else:
        above_50 = "FALSE"

    top_three = fill_list(sorted_normalized)

    entry = [row[0], row[1], row[2], above_50, top_three[0][0], top_three[1][0], top_three[2][0]]
    content_tf.append(entry)

header = {0: "Name", 1:"Category", 2:"Nb_TF_ReMap_ALL", 3:"1_most_present_above_50%", 4:"1_most_present", 5:"2_most_present", 6:"3_most_present"}
df_header = pd.DataFrame([header])
df_tf = pd.DataFrame(content_tf)

df_f = pd.concat([df_header, df_tf]).reset_index(drop=True)
df_f.to_csv("/home/mouren/Data/final_files_tokeep/tls1/biotype/exons_TF_majority_bilan.tsv",header=None,index=None,sep="\t")
```

```python
### CellLine
content_cellLines = []
for index, row in df.iterrows():
    somme = 0
    normalized = {}
    
    for cellLine in range(len(row[4].split(","))):
        normalized[row[4].split(",")[cellLine]] = int(row[5].split(",")[cellLine])/nb_peak_cellLines[row[4].split(",")[cellLine]]
        somme += int(row[5].split(",")[cellLine])/nb_peak_cellLines[row[4].split(",")[cellLine]]

    for key,value in normalized.items():
        normalized[key] = value/somme

    sorted_normalized = sorted(normalized.items(), key=lambda x: x[1], reverse=True)

    if sorted_normalized[0][1] > 0.5:
        above_50 = "TRUE"
    else:
        above_50 = "FALSE"

    top_three = fill_list(sorted_normalized)

    entry = [row[0], row[1], row[2], above_50, top_three[0][0], top_three[1][0], top_three[2][0]]
    content_cellLines.append(entry)

header = {0: "Name", 1:"Category", 2:"Nb_TF_ReMap_ALL", 3:"1_most_present_above_50%", 4:"1_most_present", 5:"2_most_present", 6:"3_most_present"}
df_header = pd.DataFrame([header])
df_cellLines = pd.DataFrame(content_cellLines)

df_f = pd.concat([df_header, df_cellLines]).reset_index(drop=True)
df_f.to_csv("/home/mouren/Data/final_files_tokeep/tls1/biotype/exons_cellLines_majority_bilan.tsv",header=None,index=None,sep="\t")
```
